```python
import pandas as pd
import numpy as np
import re
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus import thai_stopwords
import pickle 
from gensim.corpora import MmCorpus
from gensim import corpora
RANDOM_STATE = 1
```

```python
def preprocess(text):
    stopwords = list(thai_stopwords())
    read_stopwords = pd.read_csv('C:/Pream/sub_dpste_proj/dataset/add_stopwords.csv')
    add_stopwords = read_stopwords['stopword'].values.tolist()
    result = []
    str_text = str(text).replace(' ','')
    word_token = word_tokenize(str_text, engine='newmm')
    for word in word_token:
        if(word not in stopwords + add_stopwords):
            result.append(word)
    return result
```

```python
data = pd.read_csv('C:\Pream\sub_dpste_proj\dataset\DatasetLegal.csv')
str_question = data['question'].astype(str)
str_question = str_question.map(lambda x: re.sub('[,.!?*#/]', '', x))
str_answer = data['answer'].astype(str)
str_answer = str_answer.map(lambda x: re.sub('[,.!?*#/]', '', x))
```

```python
sentense_token_question = []
for i in range(len(str_answer)):
  sentense_token_question.append(str_question[i])
processed_question = []
for sentense in sentense_token_question:
  processed_question.append(preprocess(sentense))
with open('question_token.pkl', 'wb') as f:
  pickle.dump(processed_question, f)
processed_question
```

```python
sentense_token_answer = []
for i in range(len(str_answer)):
  sentense_token_answer.append(str_answer[i])
processed_answer = []
for sentense in sentense_token_answer:
  processed_answer.append(preprocess(sentense))
with open('answer_token.pkl', 'wb') as f:
  pickle.dump(processed_answer, f)
processed_answer
```

```python
list_train_data = processed_question[:5625] + processed_answer[:5625] # 5625 document 80% of all data
list_test_data = processed_question[5625:] + processed_answer[5625:] # 1406 document 20% of all data
```

```python
train_data = {'question' : processed_question[:5625],
              'answer' : processed_answer[:5625]}
test_data = {'question' : processed_question[5625:],
             'answer' : processed_answer[5625:]}

train_data = pd.DataFrame(train_data)
test_data = pd.DataFrame(test_data)
with open('train_data.pkl', 'wb') as f:
  pickle.dump(train_data, f)
with open('test_data.pkl', 'wb') as f:
  pickle.dump(test_data, f)
train_data
```

```python
id2word = corpora.Dictionary(list_train_data)
with open('id2word.pkl', 'wb') as f:
    pickle.dump(id2word, f)
```

```python
corpus = []
for text in list_train_data:
  vec = id2word.doc2bow(text)
  corpus.append(vec)
pd.Series(corpus)
with open('train_corpus.pkl', 'wb') as f:
  pickle.dump(corpus, f)
```

```python
corpus = []
for text in list_test_data:
  vec = id2word.doc2bow(text)
  corpus.append(vec)
pd.Series(corpus)
with open('test_corpus.pkl', 'wb') as f:
  pickle.dump(corpus, f)
```
